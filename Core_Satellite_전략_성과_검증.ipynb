```python
import numpy as np
import pandas as pd

df = pd.read_excel('/content/drive/MyDrive/밴치마크대비 수익률.xlsx', index_col=0, parse_dates=True)
```

```python
df.head()
```

```python
df = df.transpose()
```

```python
df
```

```python
kospi = df['코스피']
returns = kospi.pct_change()
```

```python
import matplotlib as plt

base_date = '2023-04-03'

# 기준일자에 대한 코스피 지수
base_kospi = df.loc[base_date, '코스피']

# 등락률 계산하기
df['등락률'] = (df['코스피'] - base_kospi) / base_kospi * 100
df['기준일자 등락률'] = df['수익률']
df
```

```python
!pip install matplotlib
import matplotlib.pyplot as plt
```

```python
plt.plot(df.index, df['등락률'], label='KOSPI')
plt.plot(df.index, df['기준일자 등락률'], label='MY_PORTFOLIO')

plt.xlabel('day')
plt.ylabel('rate')

plt.legend()
plt.xticks(df.index[::7])
plt.show()
```
